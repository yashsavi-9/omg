```python
from googlesearch import search
data = search('IITM',num_results=10)
all_data = []
for i in data:
  all_data.append(i)
file = open('Data Science.txt',mode = 'w')
file.writelines(all_data)
file.close()
print('Done')
```

```python
from googlesearch import search
title='AI ML'
data=search(title,num_results=10)
all_data=[]
for i in data:
    all_data.append(i+'\n')
file=open(f'{title}.txt',mode='w')
file.writelines(all_data)
file.close()
print('done')
```

```python
url='https://picsum.photos/200/300'
import requests as r
n=10
for i in range(n):
      img_data=r.get(url).content
      f=open(f'image_{i}.jpg','wb')
      f.write(img_data)
      f.close()
      from IPython import display as d
      display(d.Image(f'image_{i}.jpg',width=200,height=200))
  
```
