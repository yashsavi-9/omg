```python
#entering values in a file
from googlesearch import search
data=search('python tutorial pdf',num_results=10)
all_data=[]
for i in data:
   # print(i)
    all_data.append(i+'\n')#to send data into an empty list we are using append
    

file=open('python tutorial pdf',mode='w')
file.writelines(all_data)#to write multiple lines in a file
file.close()
print('done')
```

```python
#making file name same as the name searched
from googlesearch import search
title='python tutorial pdf'
data=search(title,num_results=10)
all_data=[]
for i in data:
  #print(i)
   all_data.append(i+'\n')#to send data into an empty list we are using append
    

file=open(f'{title}.txt',mode='w')
file.writelines(all_data)#to write multiple lines in a file
file.close
print('done')
```

```python
#using x mode to prevent overrisding and will create a new file and will return an error if a file already exists adn sill not disturb the previous data
from googlesearch import search
title='pythn'
data=search(title,num_results=10)
all_data=[]
for i in data:
  #print(i)
   all_data.append(i+'\n')#to send data into an empty list we are using append
    

file=open(f'{title}.txt',mode='x')
file.writelines(all_data)#to write multiple lines in a file
file.close
print('done')
```

```python
#apped mode is used to add data in  an existing file or if a file is not already there it will create a new file
#append mode will append 5 output every time we run this code to the file
from googlesearch import search
title='government jobs in 2025'
data=search(title,num_results=5)
all_data=[]
for i in data:
  #print(i)
   all_data.append(i+'\n')#to send data into an empty list we are using append
    

file=open(f'{title}.txt',mode='a')
file.writelines(all_data)#to write multiple lines in a file
file.write('-------new result------')
file.close
print('done')
```

```python
# how to create a copy of a file
f1=open('day_4.txt',mode='r')
all_data=f1.read()

f2=open('day_4_copy.txt','w')
f2.write(all_data)

f1.close()
f2.close()
print('done')
```
