```python
class Car:

    def __init__(self):
        self.color = 'Black'
        self.wheel = 4
        self.spare_wheel = 1
        self.airbag_support = True
        self.fuel_type = 'Petrol'
        self.car_type = 'Sedan'
        self.gear_type = 'Automatic'

   
    def auto_park(self):
        import IPython
        url = '''https://cdn.shopify.com/s/files/1/0761/9962/5012/files/parking_sensor_blog_2_480x480.gif'''
        display(IPython.display.Image(url))

    def engine_on(self):
        import pygame
        pygame.init()
        path = r'''C:/Users/paras/Downloads/engine-start-86242.mp3'''
        music = pygame.mixer.Sound(path)
        music.play()

    def speed_up(self):
        c = 10
        while c<=200:
            import time
            print(f'Speed is Increasing by {c} km/hr')
            c+= 10
            display(clear= True)
            time.sleep(0.5)

            if c >=100:
                print('You are crossing speed limit')
                break

    def drift_brake(self,drift = False):
        if drift == True:
            import pygame
            pygame.init()
            path = r'''C:/Users/paras/Downloads/mixkit-speeding-swoosh-1484.wav'''
            music = pygame.mixer.Sound(path)
            music.play()
            import IPython
            url = '''https://media.tenor.com/puPGAEN97U0AAAAM/lol-mekro.gif'''
            display(IPython.display.Image(url))

        else:
            print('Brake Applied!!')
```

```python
class ATM:
    def __init__(self):
        print('object created')
        self.pin=1111
        self.balance=2000
    def change_pin(self):
        self.pin= input('enter new password')
        print('you PIN has been changed')
    def show_balance(self):
        print(f'''your current balance is rs{self.balance}''')
    def atm_performance_status(self):
        print('ATM working fine 100%')
```
